# 小结

本章介绍了 Python 内置的几个常用的基本结构类型：
- 列表（`list`）
- 元组（`tuple`）
- 字典（`dict`）
- 集合（`set`）

剖析和演示了如何创建这些结构对象、特性以及使用说明。

在本章还新介绍 Python 的内置函数：

- 数据类型函数
    - `list()`，列表
    - `tuple()`，元组
    - `dict()`，字典
    - `set()`，集合

- 序列运算函数
    - `max()`，返回序列最大的元素
    - `min()`，返回序列最小的元素
    - `sum()`，计算序列元素求和
    - `all()`，判断序列所有元素是否为真
    - `any()`，判断序列任一元素是否为真

前面提过，数据集的数据可以用如下几个维度来描述：
- 同质（Homogeneous）或异质（inhomogeneous），数据集中的数据类型一样或不一样
- 可变（mutable）或不可变（immutable），数据集可以更改还是不允许更改
- 独一（unique）或重复（duplicated），数据集中的元素是独一或可重复
- 有序（ordered）或无序（unordered），数据集中的元素有序或无序
- 映射（mapping）或无映射（no mapping），数据集中的元素有映射或无映射

这里使用笛卡尔乘积，可以得到更多的数据结构预期：

```python
import itertools

a1 = ['同质', '异质']
a2 = ['可变', '不可变']
a3 = ['独一', '重复']      
a4 = ['有序', '无序']
a5 = ['映射', '无映射']
for i, xa in enumerate(itertools.product(a1, a2, a3, a4, a5)):
    print('结构{}：{}'.format(i, xa))
```

显然除了字符串（str）、列表（`list`）、元组（`tuple`）、 字典（`dict`）、集合（`set`）之外，还应该有很多数据结构值得去学习。在标准库`collections`中还存在一些数据结构类型值得掌握。
