# Matplotlib 图表类型

在前面示例中，使用 Matplotlib 绘制了线性图表。在 Matplotlib 库还支持许多图表类型，例如：
- 对数图
- 散点图
- 直方图
- 饼图
- 条形图
- 图像绘图

本节来演示如何绘制这些图表类型。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 对数图

下面来绘制对数图：

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# build data
x1 = np.arange(0.0, 10.0, 1)
y1 = np.exp(x1)
x2 = np.arange(0.0, 10.0, 0.1)
y2 = np.exp(1.1*x2)

# Plot...
plt.semilogy(x1, y1, '.', color='r', label='Schurmann 2005')
plt.semilogy(x2, y2, '-', color='m', label='Schurmann 2011')
plt.xlabel('$E_{C.M.}$ (MeV)')
plt.ylabel('S$_{TOT}}$ (E) (KeV b)')
plt.title('SINAP Demo B')
plt.legend(fontsize='x-small')
plt.savefig('demo02a.png')
plt.show()
```

## 散点图

散点图用来表现二个变量或者多个变量之间的关系，使用`plt.scatter()`，主要输入参数包括：
- 数据A，对应坐标轴`x` 
- 数据B，对应坐标轴`y`

```python
ages = np.random.normal(35.0, 10.0, 1000)
saleries = np.random.normal(50000, 10000, 1000)
plt.scatter(ages, saleries)
```

```python
ages = np.random.normal(35.0, 10.0, 1000)
saleries = np.random.normal(50000, 10000, 1000)
capitals = ages * saleries
plt.scatter(ages, capitals, marker='*')
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
area = np.pi * (np.random.randint(1,15,50))**2 
color = np.random.rand(50)

plt.scatter(x, y, s=area, c=color, alpha=0.5)
plt.colorbar()
```

## 直方图

使用`plt.hist()`，输出参数
- 数据
- 条状数

```python
# 以50000为中心，标准差25000的正态分布数据（10000个）。
incomes = np.random.normal(50000, 25000, 10000)
plt.hist(incomes, 61)
plt.show()
```

## 绘制饼图

调用`plt.pie()`，并传入一系列参数，包括:
- x: 数据数组
- colors: 颜色
- labels: 标签
- explode: 是否要让某一项爆炸显示（突出显示），如果是的话，程度是多少。

```python
values = [12, 55, 4, 32, 14]
colors = ['r', 'g', 'b', 'c', 'm']
explode = [0, 0, 0.2, 0, 0]
labels = ['India', 'United States', 'Russia', 'China', 'Europe']
plt.pie(values, colors=colors, labels=labels, explode=explode)
plt.title('Student Locations')
plt.show()
```

## 条形图

条状图与饼状图很类似：
- X
- height
- colors: 颜色
- tick_labels: 标签

```python
values = [12, 55, 4, 32, 14]
colors = ['r', 'g', 'b', 'c', 'm']
explode = [0, 0, 0.2, 0, 0]
labels = ['India', 'United States', 'Russia', 'China', 'Europe']
plt.bar(np.arange(len(values)), values, color=colors, tick_label=labels)
plt.show()
```

```python
values = [12, 55, 4, 32, 14]
colors = ['r', 'g', 'b', 'c', 'm']
explode = [0, 0, 0.2, 0, 0]
labels = ['India', 'United States', 'Russia', 'China', 'Europe']
plt.barh(np.arange(len(values)), values, color=colors, tick_label=labels)
plt.show()
```

```python
# bottom的使用，绘制叠加条形图
size = 5
ind = range(5)
height1 = [2,1,4,6,3]
height2 = [1,3,4,5,5]

plt.bar(ind, height=height1)
plt.bar(ind, height=height2, bottom=height1)
```

```python
# ind的使用，绘制平行条形图
size = 5
ind = np.arange(5)
height1 = [2,1,4,6,3]
height2 = [1,3,4,5,5]

plt.bar(ind, height=height1, width=0.4)
plt.bar(ind+0.4, height=height2, width=0.4)
```

```python
# bottom的使用，绘制正负对比条形图
size = 5
ind = range(5)
height1 = [2,1,4,6,3]
height2 = np.array([1,3,4,5,5])*(-1)
plt.bar(ind, height=height1, color='r')
plt.bar(ind, height=height2, color='b')
```

## boxplot

## 图像绘图

下面使用子图来绘制对个图像：

### contour

- contour(Z)    
  绘制Z的等高线。自动选择水平值   
- contour(X, Y, Z）  
  绘制X,Y和Z的等高线。X和Y数组为（x,y)平面坐标

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)
Z1 = (1-X/2 + X**5 + Y**5)* np.exp(-X**2 - Y**2)
Z2 = (1-X/2 + X**4 + Y**4)* np.exp(-X**2 - Y**2)
Z3 = (1-X/2 + X**3 + Y**3)* np.exp(-X**2 - Y**2)
Z4 = (1-X/2 + X**2 + Y**1)* np.exp(-X**2 - Y**2)
Z5 = (1-X/2 + X**2 + Y**1)* np.exp(-X**2 - Y**2)
Z6 = (1-X/2 + X**1 + Y**2)* np.exp(-X**2 - Y**2)

plt.subplot(3, 3, 1)
plt.contourf(X, Y, Z1)
plt.annotate('a', xy=(0.0, 0.0), xytext=(0.1, 0.9), xycoords='figure fraction', fontsize='large')

plt.subplot(3, 3, 2)
plt.contourf(X, Y, Z2)
plt.annotate('b', xy=(0.0, 0.0), xytext=(0.4, 0.9), xycoords='figure fraction', fontsize='large')

plt.subplot(3, 3, 3)
plt.contourf(X, Y, Z2)
plt.annotate('c', xy=(0.0, 0.0), xytext=(0.75, 0.9), xycoords='figure fraction', fontsize='large')

plt.subplot(3, 2, 3)
plt.contourf(X, Y, Z3)
plt.annotate('d', xy=(0.0, 0.0), xytext=(0.1, 0.6), xycoords='figure fraction', fontsize='large')

plt.subplot(3, 2, 4)
plt.contourf(X, Y, Z4)
plt.annotate('e', xy=(0.0, 0.0), xytext=(0.6, 0.6), xycoords='figure fraction', fontsize='large')

plt.subplot(3, 2, 5)
plt.contourf(X, Y, Z5)
plt.annotate('f', xy=(0.0, 0.0), xytext=(0.1, 0.3), xycoords='figure fraction', fontsize='large')

plt.subplot(3, 2, 6)
plt.contourf(X, Y, Z6)
plt.annotate('g', xy=(0.0, 0.0), xytext=(0.6, 0.3), xycoords='figure fraction', fontsize='large')

# export a pdf file
plt.savefig('demo04a.png')
plt.show()
```
