Matplotlib 是 Python 编程中最著名的绘图库之一，是 Python 世界的基础图形包，目的是为 Python 构建一个类似 Matlab 的绘图接口。其主要功能包括：
- 支持交互式绘图，可创建出版质量的图表
- 支持2D绘图和部分3D绘图
- 输出文件支持 png、pdf、ps 和 svg 等格式

Matplotlib 库优势在于出图快，默认设置美观。文档相当完备，提供有上百幅示例图像，复制粘贴并进行简单修改，基本上都能创建自己需要的图像。Matplotlib 库的细节很多，可以边用边学，按需使用。

本章会介绍 Matplotlib 库的使用，并给出众多常用图的实例。

## 绘图基本流程

一个绘图基本流程为，首先获取数据，并进行规制和分析；然后选择正确的图形类型把数据可视化；最后输出为图形文件。

Matplotlib 提供大量图像类型，例如常用类型如下：
- 线图
- 对数图
- 散点图
- 直方图
- 饼图
- 条形图
- 图像绘图

Matplotlib 库使用后端来输出图形图像文件，支持的图形文件格式包括：PDF、SVG、JPG、PNG、PS、BMP、GIF等

## 快速入门

使用 `matplotlib` 库的子包 `pyplot` 模块，可以调用函数来快速绘图，使用之前需要导入模块：

- 绘图
- 输出png文件

```python
import numpy as np
import matplotlib.pyplot as plt
```

使用`numpy`库的随机模块来创建一个随机数数组：

```python
# 创建一个随机数数组
y = np.random.random(1024)
```

使用 `matplotlib.pyplot` 模块的`plot()`函数，选择线图来进行数据可视化：

```python
# 绘图
plt.plot(y)
```

从上图可知，这幅 Matplotlib 图为线型图表，图表的坐标轴长度、范围、刻度、标签、线型等均为缺省值。

最后使用`plt.savefig()`函数来输出指定格式图形文件：

```python
# 输出png文件
plt.savefig('randomdemo.png')
```

`plt.savefig()`函数根据文件后缀名，使用不同后端创建指定格式图形文件。

当然也可以使用`plt.show()`函数在GUI后端予以显示。

下面把上面这些代码，保存为 Python 程序：

```python
%%writefile firstplot.py
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""Matplotlib演示 - firstplot
"""
import numpy as np
import matplotlib.pyplot as plt


def firstplot(outfile):
    """Plot a random float array
    """
    # build a random floats array
    y = np.random.random(1024)

    # plot them
    plt.plot(y.cumsum())

    # export a pdf file
    plt.savefig(outfile)
    plt.show()

    
if __name__ == '__main__':
    outfile = 'randomdemo.pdf'
    firstplot(outfile)
```

最后来运行该程序文件，并生成第一幅图形文件。

```python
%run firstplot.py
```
