# Matplotlib 绘图参数

在上一节中，创建了一幅 Matplotlib 线型图表，图表中坐标轴长度、范围、刻度、标签、线型等均使用缺省值。在 Matplotlib 绘图中，选择图表类型进行数据可视化后，可以根据自己审美来进行如下设置操作来更改 Matplotlib 绘图：
- 设置图表的线型、属性和格式化字符串
- 设置坐标轴长度和范围
- 设置刻度、标签等
- 添加图例和注解
- ...

本节介绍 Matplotlib 常用绘图参数，并介绍如何更改。

## 实例介绍

下图为中科院上海光源网站发表的一幅图像，下面使用 Matplotlib 库来演示如何生成这幅图。
![上海光源新闻线图](../images/matplotlib_sinap01.png)

从图可以看出，图像包括如下内容：
- 图表类型为线型图
- 坐标轴（x轴和y轴，axis）
- 坐标轴刻度(values)
- 刻度标签(tickers)
- 坐标轴文字 （xlabel， ylabel）
- 绘图区域(area)

由于没有真实数据，这里使用 `numpy.linspace()` 函数来仿真等间隔数组，并使用 `numpy.sin()` 函数来计算正弦值，组成`x`与`y`轴数据：

````python
import numpy as np
import matplotlib.pyplot as plt

# build data
x1 = np.linspace(-0.5*np.pi, 1.5*np.pi, 128)
y1 = np.sin(x1)
x2 = np.linspace(-0.5*np.pi, 1.5*np.pi, 256)
y2 = 1.2 * np.sin(x2)
x3 = np.linspace(-0.5*np.pi, 1.5*np.pi, 512)
y3 = 1.4 * np.sin(x3)
x4 = np.linspace(-0.5*np.pi, 1.5*np.pi, 1024)
y4 = 1.6 * np.sin(x4)
x5 = np.linspace(-0.5*np.pi, 1.5*np.pi, 2048)
y5 = 1.8 * np.sin(x5)
````

然后，调用`plt.plot()`函数绘制线型图，同时保存文件：

````python
# Plot...
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
plt.plot(x4, y4)
plt.plot(x5, y5)

# export a pdf file
plt.savefig('demo01a.png')
plt.show()
````

上图为使用 Matplotlib 缺省值得到的结果图形。

线性图表用来表现二个变量之间的关系。使用内置函数`help()`打印`plt.plot()`帮助，

````python
help(plt.plot)
````

可以看出 `plt.plot()` 的参数众多，使用方法灵活多变。传入参数有如下情况：
- `plot(x, y)`， 绘制`x`与`y`的线型图，使用缺省参数；
- `plot(x, y, 'bo')`， 绘制`x`与`y`的线型图，使用指定参数；
- `plot(y)`，一个参数，则`x`缺省为索引数组；
- `plot(y, 'r+')`，`x`缺省为索引数组，使用指定参数

如果`x`或`y`是多维矩阵，则相当于画多个折线图。也可以一次绘制多个点或线：
```python
plot(x1, y1, x2, y2, 'g-')
plot(x1, y1, 'g^', x2, y2, 'g-')
```


参数有fmt形式和kwargs形式
 'g^-'   VS    color='g', marker='^', linestyle='-'


## 设置线型属性

线型属性包括如下：
- 线条风格（linestyle）
- 线条标记（marker）
- 线宽（linewidth）
- 颜色（color）
- 标签（label）
- 透明度（alpha）

### 线条风格（linestyle）

可以通过位置参数或使用关键字参数`linestype`来指定线条风格：
```python
plot(x, y, '--')
plot(y, '-')
plot(x, y, linestyle='--')
```

线条风格对应字符及描述如下所示：

| 线条风格 |  描述| 线条风格 |  描述|
|:---------|:------|:---------|:------|
|``'-'``  | 实线  | ``'--'`` | 破折线   | 
|``'-.'`` | 点划线 | ``':'`` | 虚线 |
|``''``  |  无    |   |    |

````python
x1 = np.linspace(-0.5*np.pi, 1.5*np.pi, 64)
y1 = np.sin(x1)
plt.plot(x1, y1, '--')
````

### 线条标记（marker）

可以通过位置参数来设置线条标记，或者使用关键字参数`marker`来指定线条标记：
```python
plot(x, y, 'o')
plot(x, y, markee='.')
```

如果指定了`marker`参数，则会绘制出散点图。线条标记对应字符及描述如下所示：

| 线条标记 |  描述| 线条标记 |  描述|
|:---------|:------|:---------|:------|
|``'.'``  | 点    | ``','``  |像素 |
|``'o'``  | 圆圈  | ``''`` | 无 |
| ``'v'`` | 朝下三角形 |``'^'``  | 朝上三角形 |
|``'<'``  |朝左三角形  |``'>'``  | 朝右三角形 |
|``'1'``  | 朝下三角  |``'2'``  | 朝上三角 |
|``'3'``  | 朝左三角  |``'4'``  | 朝右三角|
|``'s'``  | 正方形   |``'p'``    | 五边形 |
|``'*'``  |星号 | ``'h'`` | 六边形1 |
|``'H'``  | 六边形2 |``'+'`` | 加号 |
|``'x'``  | `X`  |``'D'``  | 菱形 |
|``'d'``  | 小菱形 | `'`<code>&#124;</code>`'`     |  竖线 |
|``'_'``  |水平线 |  |   |


````python
x1 = np.linspace(-0.5*np.pi, 1.5*np.pi, 64)
y1 = np.sin(x1)
plt.plot(x1, y1, marker='.', linestyle='-')
````

###  线宽（linewidth）

使用关键字`linewidth`来指定线宽，其值为浮点数

````python
x1 = np.linspace(-0.5*np.pi, 1.5*np.pi, 64)
y1 = np.sin(x1)
plt.plot(x1, y1, linestyle='--', linewidth=5)
````

### 颜色（color）

可以使用关键字参数`color`来指定颜色。颜色可以使用名字来指定，例如一些常用颜色名字如下所示：

| 颜色 |  描述| 颜色 |  描述|
|:---------|:------|:---------|:------|
|``'b'``  | 蓝色   | ``'g'``  | 绿色 |
|``'r'``  | 红色  | ``'c'`` | 青色 |
| ``'m'`` |  |``'y'``  | 黄色 |
|``'k'``  | 洋红色   |``'w'``  | 白色 |

也可以使用 HTML 十六进制字符串或 RGB 元组。例如：
```python
bcolor = '#eeefff'
fcolor = (0.3, 0.3, 0.4)
```

````python
x1 = np.linspace(-0.5*np.pi, 1.5*np.pi, 64)
y1 = np.sin(x1)
plt.plot(x1, y1, '+', color='red')
````

### 标签（label）

任意字符串，为图例设置标签值:

````python
x1 = np.linspace(-0.5*np.pi, 1.5*np.pi, 64)
y1 = np.sin(x1)
plt.plot(x1, y1, 'k-', label='sin')
````

### 透明度（alpha）

浮点值，用来设置颜色，并不是所有后端都支持。

下面是设置线下属性的示例：

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# build data
x1 = np.linspace(-0.5*np.pi, 1.5*np.pi, 128)
y1 = 2 * np.sin(x1)
x2 = np.linspace(-0.5*np.pi, 1.5*np.pi, 256)
y2 = 1.7 * np.sin(x2)
x3 = np.linspace(-0.5*np.pi, 1.5*np.pi, 512)
y3 = 1.4 * np.sin(x3)
x4 = np.linspace(-0.5*np.pi, 1.5*np.pi, 1024)
y4 = 1.2 * np.sin(x4)
x5 = np.linspace(-0.5*np.pi, 1.5*np.pi, 2048)
y5 = 1.0 * np.sin(x5)

# Plot...
plt.plot(x1, y1, '-', linewidth=1, color='b', label='G-FeCow at open circuit')
plt.plot(x2, y2, '--', linewidth=1, color='g', label='G-FeCow at $+1.4$ V')
plt.plot(x3, y3, '-.', linewidth=1, color='r', label='A-FeCow at open circuit')
plt.plot(x4, y4, ':', linewidth=1, color='c', label='A-FeCow at $+1.4$ V')
plt.plot(x5, y5, '-', linewidth=1.2, color='m', label='WO$_3$ ref')

# export a pdf file
plt.savefig('demo01b.png')
plt.show()
````

## 设置注解

使用`plt.annotate()`函数来在图中添加注解，主要参数包括：
- 标签（label）  
  注解字符串
- 位置（xy）
- 坐标（xycoords）
- 文本位置（xytext ）
- 箭头（arrowprops）

在下面代码中，添加了两个注解：

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# build data
x1 = np.linspace(-0.5*np.pi, 1.5*np.pi, 128)
y1 = 2 * np.sin(x1)
x2 = np.linspace(-0.5*np.pi, 1.5*np.pi, 256)
y2 = 1.7 * np.sin(x2)
x3 = np.linspace(-0.5*np.pi, 1.5*np.pi, 512)
y3 = 1.4 * np.sin(x3)
x4 = np.linspace(-0.5*np.pi, 1.5*np.pi, 1024)
y4 = 1.2 * np.sin(x4)
x5 = np.linspace(-0.5*np.pi, 1.5*np.pi, 2048)
y5 = 1.0 * np.sin(x5)

# Plot...
plt.figure(figsize=(6,4),dpi=600)
plt.plot(x1, y1, '-', linewidth=1, color='b', label='G-FeCow at open circuit')
plt.plot(x2, y2, '--', linewidth=1, color='g', label='G-FeCow at $+1.4$ V')
plt.plot(x3, y3, '-.', linewidth=1, color='r', label='A-FeCow at open circuit')
plt.plot(x4, y4, ':', linewidth=1, color='c', label='A-FeCow at $+1.4$ V')
plt.plot(x5, y5, '-', linewidth=1, color='m', label='WO$_3$ ref')
plt.xlim(0, 1.5*np.pi)
plt.xticks([-np.pi/2, 0, np.pi/2, np.pi, np.pi*3/2],
           [r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$'])
#plt.yticks([-2, 1, 0, 1, 2])
plt.yticks([])
plt.annotate('Peak', xy=(np.pi/2, 2), xytext=(0, 1.5), arrowprops=dict(arrowstyle='->'))
plt.annotate('C', xy=(0.0, 0.0), xytext=(0.03, 0.9), xycoords='figure fraction', fontsize='xx-large')
plt.xlabel('Energy (eV)')
plt.ylabel('WL$_3$-edge XANES Intensity (A.U.)')
plt.title('SINAP Demo A')
# plt.legend()
plt.legend(fontsize='x-small')
# export a pdf file
plt.savefig('demo01f.png')
plt.show()
````
