# Matplotlib 中文显示

在 Matplotlib 版本中，中文显示常常会出现乱码问题。有如下解决方案：

- 在代码中指定字体配置：

```python
import matplotlib as mpl

fname = '/usr/share/fonts/cjkuni-uming/uming.ttc'
zhfont = mpl.font_manager.FontProperties(fname=fname)
plt.title(u'$sin$函数', fontproperties=zhfont)
```

- 动态设置
  - 拷贝 windwos 系统中 fonts 目录下的`simhei.ttf` 拷贝到 `site-packages/matplotlib/mpl-data/fonts/ttf`目录中，
  - 删除~/.cache/matplotlib的缓冲目录
  - 编写代码

```python
import matplotlib as mpl

# 指定默认字体  
matplotlib.rcParams['font.family']='sans-serif'  
matplotlib.rcParams['font.sans-serif'] = ['SimHei’]   
# 解决负号'-'显示为方块的问题  
matplotlib.rcParams['axes.unicode_minus'] = False 
plt.title(u'$sin$函数')
```

- `rc`文件中指定缺省字体  
  修改 matplotlib 配置文件

## 指定字体配置

下面示例为在 Linux 系统下，通过在代码中指定字体配置来解决中文显示问题：

````python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# 在代码中指定字体配置
# 设置中文字体
# fname='/usr/share/fonts/cjkuni-uming/uming.ttc'
# Windows 系统
fname='C:\Windows\Fonts\SimHei.ttf'
# fname='/usr/share/fonts/cjkuni-uming/uming.ttc'
zhfont = mpl.font_manager.FontProperties(fname=fname)

x = np.linspace(0, 2*np.pi, 128)
y = np.sin(x)

# Plot...
plt.plot(x, y)

plt.xlabel(u'X 轴', fontproperties=zhfont)
plt.ylabel(u'Y 轴', fontproperties=zhfont)
plt.title(u'$sin$函数', fontproperties=zhfont)

# export a pdf file
plt.savefig('demo06a.png')
plt.show()
````

## 动态设置

下面示例为动态设置方法解决中文显示问题：

````python
#-*-coding:utf-8-*-
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# 指定默认字体  
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['font.family']='sans-serif'  
# 解决负号'-'显示为方块的问题  
mpl.rcParams['axes.unicode_minus'] = False 
x = np.linspace(0, 2*np.pi, 128)
y = np.sin(x)

# Plot...
plt.plot(x, y)
plt.xticks([-np.pi/2, 0, np.pi/2, np.pi, np.pi*3/2],
           [r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$'])
plt.xlabel('X 轴')
plt.ylabel('Y 轴')
plt.title(u'$sin$ 函数')

# export a pdf file
plt.savefig('demo06b.png')
plt.show()
````
